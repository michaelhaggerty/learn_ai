```python
import requests
import json

# URL and Authorization Key
url = "https://api.webselfstorage.com/v3/location/1038105"
headers = {
    "Authorization": "Bearer API_KEY",  # Replace 'KEY' with your actual Bearer token
    "Accept": "application/json"
}

# Make the GET request
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse JSON data from the response
    data = response.json()
    # Pretty print the JSON data
    #print(json.dumps(data, indent=4))
else:
    print(f"Failed to retrieve data: {response.status_code}")


# Access the 'units' key which is a list of units
units = data['location']['units']

# Create a dictionary with unit size and vacant units
unit_size_vacant = [
    {'unitSize': unit['unitSize'], 'vacantUnits': unit['vacantUnits']}
    for unit in units
]

# Print the dictionary
print(unit_size_vacant)
```
